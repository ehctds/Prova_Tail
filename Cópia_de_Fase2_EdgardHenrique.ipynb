# Questão 1: Python Básico
---



**Descrição:** 
\
Strings são variáveis que recebem valores textuais. Em python, uma string pode ser visto como uma lista. Digamos que você foi contratado por uma grande empresa de busca de palavras, e o seu chefe te passa a tarefa de verificar se certas palavras são o inverso da outra (ex: pedro e ordep ou mesa e asem). Você, como uma pessoa esperta, resolve automatizar um script que verífica se as palavras são ou não iguais ao seu inverso.

\\
**Tarefa:**
\
Implemente uma script em Python para veríficar se duas strings são inversas.

\\
Referências:

[1] [youtube.com/watch?v=1OQ9e5-OiKM](https://youtube.com/watch?v=1OQ9e5-OiKM)

```python
# Seu código aqui
var1 = input("qual é a primeira palavra ?").lower() #aqui o usuario realiza o input da primeira palavra que será utilizada na comparação
var2 = input("qual é a segunda palavra ?").lower() #aqui o usuario realiza o input da segunda palavra que será utilizada na comparação

if var2 == var1[::-1]:
    print("as strings são inversas") 
else:
    print("as strings não são inversas") 
#Aqui será feito a comparação entre as duas strings, utilizando o comando if e else do python 

#Como o python não sabe identificar os diferentes caracteres(maisculo ou minusculo) da mesma palavra, decidi por colocar o comando lower, para diminuir as chances de erro do usuario.
```

### 5.1 - Explique com suas palavras o *trade-off* entre em *Exploration* e *Exploitation* no contexto de Aprendizagem por Reforço.
---

De forma geral, o método de “Exploitation” funciona utilizando o conceito de “Rewards” e estimativas positivas da variável “Action” até ser possível de identificar um valor estimado possivelmente ideal, no entanto, após achar este valor, o método para de utilizar as demais possibilidades, causando muitas incertezas para o programador se o valor encontrado realmente é o ideal ou não, afinal é muito provável que exista outro valor que se sobressaia nas diversas variáveis que foram consideradas tão boas quanto. 
 
Ao contrário do método de “Exploitation”, o método de “Exploration” continuará tentando sempre melhorar a variável “Action” também utilizando o conceito de "Rewards" mas não se importara com o quão bom já esteja, ou seja, estará sempre em constante evolução. Um grande exemplo disso pode ser um sistema de recomendação que sempre mira em melhorar a satisfação dos clientes a longo prazo, tentando entender os gostos antigos do cliente e trazendo sempre conteúdos mais relevantes para ele. Isto não funcionário no método de “Exploitatation” pois não haveria um foco na diversidade de gosto do cliente ao longo do tempo, o sistema entendera apenas 1 dos focos de gostos e trará conteúdos sempre parecidos.

Como conclusão, podemos dizer que o método de “Exploitation” funciona melhor no curto prazo, trazendo informações relevantes de forma rápida, mas não tão precisas quanto no método de “Exploration” que precisara de um tempo muito maior para chegar ao ponto ideal de ações de melhoria.  Ou seja, o ideal é que haja um balanceamento entre os dois métodos para uma maior otimização.
